## Preparing data

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#loading dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

#visualizing dataset
X_pd = pd.DataFrame(X)
y_pd = pd.DataFrame(y)
pd.plotting.scatter_matrix(X_pd, alpha=0.2, c=y_pd)
```

^ Feature 3 against 2 gives better separation.

```python
#Reassign dataset
X_n = X[:,[2,3]]
type(X_n)
```

## KNN implementation

```python
#data inquiry
print('Input sample data:')
X_p = np.array([0,0])
X_p[0] = float(input())
X_p[1] = float(input())
print('Gain sample data: {}\n'.format(X_p))

print('Input K neighbours count:')
K = int(input())
print('Gain K count: {}\n'.format(K))
```

```python
X_nn.shape[0]
X_nn[10]
```

```python
#calculating:
X_nn = X_n
y_nn = y
min_X = np.array([X_nn[0],0,9999999]) #cords,index,distance
found_y = []

if (K>3):
    #looking for neighbors
    for k in range(K):
        for i in range(X_nn.shape[0]):
            dist = np.absolute( np.sqrt( (X_nn[i,0]-X_p[0])**2 + (X_nn[i,1]-X_p[1])**2))
            if ( min_X[2] >= dist):
                    min_X = np.array([X_p,i,dist])
        found_y = np.append(found_y, int(y_nn[min_X[1]]))
        print(min_X[1])
        X_nn = np.delete(X_nn, min_X[1], 0)
        
    #voting for class assignment
    counts = np.bincount(found_y)
    y_p = np.argmax(counts)

print('Assigned class is: {}'.format(y_p))    
```
