```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
females_titanic = titanic[titanic['Sex']=='female']
females35_titanic = females_titanic[females_titanic['Age']>35]
females35_titanic.to_csv('females35_survived')

check = pd.read_csv('females35_survived')
check.describe()
```

```python
age_fare = titanic[['Age','Fare']]
age_fare.sort_values(by='Age')
```

```python
x = np.array(age_fare['Age'])
y = np.array(age_fare['Fare'])
plt.plot(x,y,'o')

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Dependence of Fare out of Age')
```

```python
males = titanic[titanic['Sex']=='male']
males_count =  np.array(males.shape)[0]
print(males_count)

females = titanic[titanic['Sex']=='female']
females_count =  np.array(females.shape)[0]
print(females_count)

x1 = ['males','females']
y1 = [males_count,females_count]
plt.plot(x1,y1)

plt.xlabel('Passengers by their sex')
plt.ylabel('Passengers count')
plt.title('Passengers')
```

```python
k = np.array(titanic.shape)[0]
```

```python
i = 0
count = 0
fare_val = 100
for i in range(k):
    if np.array(titanic['Fare'])[i] < fare_val:
        count = count+1

percent = k/100
percentage = count/percent

print('There were ',count,' of people that had a fare of less than', fare_val)
print('Its ',percentage,'% of all amount')
```

```python
i = 0
count = 0
survived = 0
for i in range(k):
    if np.array(titanic['Age'])[i] <= 10:
        count = count+1
        if np.array(titanic['Survived'])[i] == 1:
            survived = survived+1

print('There were ',count,' children')
print(survived,' of them survived')
```
